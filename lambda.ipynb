```python
def square_numbers(nums):
    squared_list = []
    for i in nums:
        squared_list.append(i ** 2)

    return squared_list

nums = [1,2,3,4,5]
result = square_numbers(nums)
print(result)
```

Reduce and Filter

```python
from functools import reduce

students = [
    {"name": "hanvika", "scores": [85, 90, 79]},
    {"name": "shreshta", "scores": [88, 92, 89]},
    {"name": "shanvi", "scores": [79, 78, 80]},
    {"name": "shiva", "scores": [90, 81, 88]},
]

filtered_students = list(
    filter(
        lambda s: sum(s["scores"]) / len(s["scores"]) > 80,
        students
    )
)

total_average = reduce(
    lambda total, s: total + sum(s["scores"]) / len(s["scores"]),
    filtered_students,
    0
    
)

print(total_average)
```
