```python
class Dog:
 
    def __init__(self,name,breed,age):
    
        self.name = name
        self.breed = breed
        self.age = age

    def bark(self):
        print(f"{self.name} barks: Bowww.....Bowww")

    def show_info(self):
        print(f"Dog's Info: I am{self.name}, my breed is {self.breed}, and {self.age} years old")

    def my_age(self):
        self.age += 1
        print(f"My name {self.name}! Now {self.age} years old")
```

```python
pug = Dog("Chers", "Pug", 6)
```

```python
pug.show_info()
```

```python
pug.bark()
```

```python
class Tree:

  def __init__(self, species, height, age, has_fruits =False):

    self.species = species
    self.height = height
    self.age = age
    self.has_fruits = has_fruits

  def describe_tree(self):
    status_fruites = "having fruites" if self.has_fruits else "Not having fruites yet"
    print(f" This is a {self.species} tree, {self.height}m tall, {self.age} years old, {status_fruites}")

  def having_fruites(self):
    if self.age > 4:
      self.has_fruits = True
      print(f"The {self.species} tree is having fruites")
    else:
      print(f"This Tree is {self.age} much younger to having fruites")
```

```python
mango_tree = Tree("Mango", 5, 6)
```

```python
mango_tree.describe_tree()
```

```python
class Movie:

    def __init__(self, title, genre, year, hero):
        self.title = title
        self.genre = genre
        self.year = year
        self.hero = hero

    def Name(self):
        print(f"This Movie is {self.title}, genre is {self.genre}, released in {self.year}, hero is {self.hero}")
              
    
    def is_old_movie(self):
        if self.year < 2003:
            print(f"{self.title} is an old movie")
        else:
            print(f"{self.title} is a new movie")

    def movie_type(self):
        if self.genre == "Comedy":
            print(f"This is a Comedy Movie")
        elif self.genre == "Action":
            print(f"This is a Action Movie")
        elif self.genre == "Horror":
            print(f"This is a Horror Movie")
        else:
            print(f"This is a Emotional Movie")
```

```python
movie1 = Movie("Pushpa", "Action", 2025, "Allu Arjun")
```

```python
movie1.movie_type()
```

```python
movie1.is_old_movie()
movie1.Name()
```
